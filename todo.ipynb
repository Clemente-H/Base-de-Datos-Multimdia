```python
import torch
from transformers import BertTokenizer, BertForQuestionAnswering

# Cargar el modelo pre-entrenado BERT
tokenizer = BertTokenizer.from_pretrained("dccuchile/bert-base-spanish-wwm-cased")
model = BertForQuestionAnswering.from_pretrained("dccuchile/bert-base-spanish-wwm-cased")
```

```python
p1 = "videojuegos"
p2 = "Pokedex"
p3 = "Steam"
```

```python
# Crear la pregunta utilizando las palabras de entrada "X, Y, Z"
question = "¿Cuál es la palabra que relaciona las palabras " + p1 +"," + p2 + "" + p3 + "?"

# Codificar la pregunta y el texto de referencia (en este caso, vacío)
encoded_question = tokenizer.encode(question, add_special_tokens=True)
encoded_text = tokenizer.encode("", add_special_tokens=True)
```

```python
# Ejecutar el modelo para generar una respuesta
inputs = torch.tensor([encoded_question, encoded_text])
start_scores, end_scores = model(inputs)

# Obtener la respuesta
answer_start = torch.argmax(start_scores)
answer_end = torch.argmax(end_scores)
answer = tokenizer.decode(encoded_question[answer_start:answer_end+1])

# Imprimir respuesta
print("Respuesta:", answer)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import AgglomerativeClustering
import numpy as np

embedder = SentenceTransformer('symanto/sn-xlm-roberta-base-snli-mnli-anli-xnli')
```

```python
corpus_embeddings = embedder.encode(corpus)

# Normalize the embeddings to unit length
corpus_embeddings = corpus_embeddings /  np.linalg.norm(corpus_embeddings, axis=1, keepdims=True)
```

```python
# Perform kmean clustering
clustering_model = AgglomerativeClustering(n_clusters=None, distance_threshold=1.5) #, affinity='cosine', linkage='average', distance_threshold=0.4)
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_

clustered_sentences = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []

    clustered_sentences[cluster_id].append(corpus[sentence_id])
```

```python
for i, cluster in clustered_sentences.items():
    print("Cluster ", i+1)
    print(cluster)
    print("")
```

```python
import pandas as pd
import gensim
# Import and download stopwords from NLTK.
from nltk.corpus import stopwords
from nltk import download
download('stopwords')  # Download stopwords list.
stop_words = stopwords.words('spanish')
def preprocess(sentence):
    return [w for w in sentence.lower().split() if w not in stop_words]
```

```python
import gensim.downloader as api
from gensim.models import Word2Vec, KeyedVectors
model = KeyedVectors.load_word2vec_format('http://dcc.uchile.cl/~jperez/word-embeddings/fasttext-sbwc.vec.gz', binary=False)
```

```python
#http://cs.famaf.unc.edu.ar/~ccardellino/SBWCE/SBW-vectors-300-min5.bin.gz
```

```python
def distancia_similitud(document1,document2,n_palabras):    
    # Identificar las palabras con mayor contribución en la distancia
    contributing_words = {}
    for word in set(document1 + document2):
        contributing_words[word] = abs(model.wmdistance([word],document1) - model.wmdistance([word],document2))
    
    # Ordenar las palabras por contribución a la distancia
    sorted_contributing_words = sorted(contributing_words.items(), key=lambda x: x[1], reverse=True)
    return sorted_contributing_words[:n_palabras],sorted_contributing_words[n_palabras:]
```
